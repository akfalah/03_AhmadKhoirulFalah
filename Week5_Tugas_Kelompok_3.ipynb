# Kelompok 3 : Gol Darah dan TTD

---



```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import drive

drive.mount('/content/drive')
```

```python
# @title Load Image
path = '/content/drive/MyDrive/pcvk/KTP.jpeg'
img = cv.imread(path)
cv2_imshow(img)
```

```python
# @title Melakukan masking untuk mengambil Golongan darah dan TTD
# TTD
# x1:600 y1=400 x2=765 y2=490
# GolDar
# x1=380 y1=170 x2=545 y2=185

# Membuat bidang masking
black_img = np.zeros(img.shape, dtype=np.uint8)
cv.rectangle(black_img, (380, 165), (545, 185), (255, 255, 255), -1)
cv.rectangle(black_img, (600, 400), (765, 490), (255, 255, 255), -1)

# Melakukan masking dengan operator AND
masked_img = cv.bitwise_and(img, black_img)

# Menampilkan gambar
concatenated_img = cv.hconcat([img, masked_img])
cv2_imshow(concatenated_img)
```
